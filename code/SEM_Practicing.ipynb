```
df = pd.read_csv("SEM.csv")
```

# measurement model
JS =~ JS1 + JS2 + JS3 + JS4 + JS5
EP =~ EP1 + EP2 + EP3 + EP4
OC =~ OC1 + OC2 + OC3 + OC4
AC =~ AC1 + AC2 + AC3 + AC4
SI =~ SI1 + SI2 + SI3 + SI4
# regressions
JS ~ EP + AC
OC ~ EP + AC +JS
SI ~ JS + OC
# residual correlations
EP ~~ AC

```
print(df.head())
```

```
df
```

```
de = '# measurement model\nJS =~ JS1 + JS2 + JS3 + JS4 + JS5\nEP =~ EP1 + EP2 + EP3 + EP4\nOC =~ OC1 + OC2 + OC3 + OC4\nAC =~ AC1 + AC2 + AC3 + AC4\nSI =~ SI1 + SI2 + SI3 + SI4\n# regressions\nJS ~ EP + AC\nOC ~ EP + AC +JS\nSI ~ JS + OC\n# residual correlations\nEP ~~ AC'
```

```
print(de)
```

```
mod = Model(de)
```

```
res_opt = mod.fit(df)
estimates = mod.inspect()
```

```
print(res_opt)
```

```
print(estimates)
```

```
g = semopy.semplot(mod, 'model.png')
g
```

```
stats = semopy.calc_stats(mod)
print(stats.T.round(3))
```
